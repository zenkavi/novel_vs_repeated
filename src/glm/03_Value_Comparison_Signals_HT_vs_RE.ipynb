# Model: "value_parametric_by_type" 

Uses the `value_parametric_by_type` model with separate parametric value
modulators for hypertrained (HT) and regular exposure (RE) stimuli.

Runs the analysis across all three scanning sessions (weeks 1, 2, 3) to
track how HT and RE value signals evolve with training.

Contrasts of interest:
  - stim_value_par_HT: value modulation for hypertrained stimuli
  - stim_value_par_RE: value modulation for regular exposure stimuli
  - stim_value_par_HT_minus_RE: HT > RE value signal (Antonio's prediction for lOFC)

```python
import os
import sys
sys.path.insert(0, '.')
 
BIDS_PATH = '/hopper/groups/enkavilab/data/novel-vs-repeated/fmri/bids'
OUTPUT_DIR = '/hopper/groups/enkavilab/users/zenkavi/novel_vs_repeated/outputs/glm'
 
SUBJECTS = ['601', '609', '611', '619', '621', '629']
SESSIONS = ['01', '02', '03']
MNUM = 'value_parametric_by_type'
SPACE = 'MNI152NLin2009cAsym_res-2'
TASKS = ['yesNo', 'binaryChoice']
 
MODEL_VARIANT = 'rt_duration_plus_mod'
 
# GLM parameters
HRF_MODEL = 'spm'
DRIFT_MODEL = 'cosine'
NOISE_MODEL = 'ar1'
SMOOTHING_FWHM = 5
SCRUB_THRESH = 0.5
 
# Report parameters
T_THRESHOLD = 3.0
CLUSTER_THRESHOLD = 10
ALPHA = 0.05
 
# Contrasts to analyze at group level
CONTRAST_IDS = [
    'stim_value_par_HT',
    'stim_value_par_RE',
    'stim_value_par_HT_minus_RE',
]
```

## Generate level 1 reports and design matrices (all sessions)

Review these to check that HT and RE regressors are well separated
and that VIFs are reasonable.

```python
from level1_helpers import generate_report
 
for session in SESSIONS:
    for subnum in SUBJECTS:
        print(f"\n{'='*60}")
        print(f"sub-{subnum} ses-{session} | {MODEL_VARIANT} ({MNUM})")
        print(f"{'='*60}")
 
        try:
            report_path, design_matrices = generate_report(
                subnum=subnum,
                session=session,
                data_path=BIDS_PATH,
                model_variant=MODEL_VARIANT,
                mnum=MNUM,
                space=SPACE,
                hrf_model=HRF_MODEL,
                drift_model=DRIFT_MODEL,
                scrub_thresh=SCRUB_THRESH,
                output_dir=OUTPUT_DIR,
            )
            print(f"  Saved {len(design_matrices)} design matrices and report.")
        except Exception as e:
            print(f"  FAILED: {e}")
            import traceback
            traceback.print_exc()
 
```

## Paired session comparisons (Week 3 vs Week 1)

changes from week 1 to week 3.
