# Model: "value_parametric" (for week 3 lOFC value encoding analysis)

Same `value_parametric` model as notebook `01_Value_Comparison_Signals_W1.ipynb`, but fitted to ses-03 (week 3, the last scanning session). After fitting, we compare ses-03 to ses-01 to test whether value signals (particularly in lOFC) increase with training.

## Generate level 1 reports and design matrices (ses-03)

```python
import os
import sys
sys.path.insert(0, '.')
 
BIDS_PATH = '/hopper/groups/enkavilab/data/novel-vs-repeated/fmri/bids'
OUTPUT_DIR = '/hopper/groups/enkavilab/users/zenkavi/novel_vs_repeated/outputs/glm'
 
SUBJECTS = ['601', '609', '611', '619', '621', '629']
SESSION = '03'
MNUM = 'value_parametric'
SPACE = 'MNI152NLin2009cAsym_res-2'
TASKS = ['yesNo', 'binaryChoice']
CONTRAST_ID = 'stim_value_par'
 
MODEL_VARIANTS = [
    'rt_in_duration',
    'rt_duration_plus_mod',
]
 
# GLM parameters (same as W1)
HRF_MODEL = 'spm'
DRIFT_MODEL = 'cosine'
NOISE_MODEL = 'ar1'
SMOOTHING_FWHM = 5
SCRUB_THRESH = 0.5
 
# Report parameters
T_THRESHOLD = 3.0
CLUSTER_THRESHOLD = 10
ALPHA = 0.05
 
# %% Step 1: Generate level 1 reports and design matrices (ses-03)
#
# This saves design matrices, VIF reports, and correlation matrices for
# all subjects and model variants. Review these before running GLMs.
 
from level1_helpers import generate_report
 
for variant in MODEL_VARIANTS:
    for subnum in SUBJECTS:
        print(f"\n{'='*60}")
        print(f"sub-{subnum} ses-{SESSION} | {variant}")
        print(f"{'='*60}")
 
        try:
            report_path, design_matrices = generate_report(
                subnum=subnum,
                session=SESSION,
                data_path=BIDS_PATH,
                model_variant=variant,
                mnum=MNUM,
                space=SPACE,
                hrf_model=HRF_MODEL,
                drift_model=DRIFT_MODEL,
                scrub_thresh=SCRUB_THRESH,
                output_dir=OUTPUT_DIR,
            )
            print(f"  Saved {len(design_matrices)} design matrices and report.")
        except Exception as e:
            print(f"  FAILED: {e}")
            import traceback
            traceback.print_exc()
 
```

## Run level 1 GLMs (ses-03)

```python
# Fits GLMs and saves contrast maps (effect_size, t-stat, z-stat) for
# all subjects x tasks x variants.
 
from level1_helpers import run_level1_pipeline
 
for variant in MODEL_VARIANTS:
    for task in TASKS:
        for subnum in SUBJECTS:
            print(f"\n{'='*60}")
            print(f"sub-{subnum} ses-{SESSION} task-{task} | {variant}")
            print(f"{'='*60}")
 
            try:
                run_level1_pipeline(
                    subnum=subnum,
                    session=SESSION,
                    task=task,
                    data_path=BIDS_PATH,
                    dm_dir=OUTPUT_DIR,
                    output_dir=OUTPUT_DIR,
                    mnum=MNUM,
                    model_variant=variant,
                    space=SPACE,
                    noise_model=NOISE_MODEL,
                    hrf_model=HRF_MODEL,
                    drift_model=DRIFT_MODEL,
                    smoothing_fwhm=SMOOTHING_FWHM,
                )
            except Exception as e:
                print(f"  FAILED: {e}")
                import traceback
                traceback.print_exc()
```

# Week 3 vs Week 1 comparison

Paired t-test: ses-03 minus ses-01 effect_size maps.
Tests whether the stim_value_par signal increases from week 1 to week 3. This step requires that level 1 GLMs for BOTH ses-01 and ses-03 have already been run (ses-01 from notebook 01, ses-03 from steps above).

Two analyses:  
    a. Whole-brain paired t-test (saves difference tmap)  
    b. ROI-based paired t-test with Bonferroni correction  

```python
SESSIONS_PAIR = ('01', '03')  # (baseline, follow-up)
```

## a: Whole-brain paired t-test

```python
from level2_helpers import run_group_paired_ttest
 
for variant in MODEL_VARIANTS:
    for task in TASKS:
        print(f"\n{'='*60}")
        print(f"Paired t-test: ses-{SESSIONS_PAIR[1]} - ses-{SESSIONS_PAIR[0]} | "
              f"task-{task} {CONTRAST_ID} | {variant}")
        print(f"{'='*60}")
 
        try:
            run_group_paired_ttest(
                subjects=SUBJECTS,
                sessions=SESSIONS_PAIR,
                task=task,
                contrast_id=CONTRAST_ID,
                output_dir=OUTPUT_DIR,
                mnum=MNUM,
                model_variant=variant,
                space=SPACE,
            )
        except Exception as e:
            print(f"  FAILED: {e}")
            import traceback
            traceback.print_exc()
 
```

## b: ROI-based paired t-test with Bonferroni correction

```python
from level2_helpers import extract_roi_betas_paired, roi_paired_ttest_table
 
for variant in MODEL_VARIANTS:
    for task in TASKS:
        print(f"\n{'='*60}")
        print(f"ROI paired test: ses-{SESSIONS_PAIR[1]} - ses-{SESSIONS_PAIR[0]} | "
              f"task-{task} {CONTRAST_ID} | {variant}")
        print(f"{'='*60}")
 
        try:
            paired_betas = extract_roi_betas_paired(
                subjects=SUBJECTS,
                sessions=SESSIONS_PAIR,
                task=task,
                contrast_id=CONTRAST_ID,
                output_dir=OUTPUT_DIR,
                mnum=MNUM,
                model_variant=variant,
                space=SPACE,
            )
 
            ttest_results = roi_paired_ttest_table(paired_betas, alpha=ALPHA)
 
            print(f"\n  Results (ses-{SESSIONS_PAIR[1]} - ses-{SESSIONS_PAIR[0]}):")
            print(ttest_results[[
                'coord_label', 'mean_beta_ses_a', 'mean_beta_ses_b',
                'mean_diff', 't_stat', 'p_uncorr', 'p_bonf', 'significant',
            ]].to_string(index=False))
 
            # Save results table
            from level2_helpers import get_group_dir, get_group_map_prefix
            group_dir = get_group_dir(OUTPUT_DIR, MNUM, variant, SESSIONS_PAIR[1])
            prefix = get_group_map_prefix(task, SPACE, MNUM, variant)
            csv_path = os.path.join(
                group_dir,
                f'{prefix}_{CONTRAST_ID}_ses{SESSIONS_PAIR[1]}_minus_ses{SESSIONS_PAIR[0]}_roi_paired_ttest.csv'
            )
            ttest_results.to_csv(csv_path, index=False)
            print(f"\n  Saved: {csv_path}")
 
        except Exception as e:
            print(f"  FAILED: {e}")
            import traceback
            traceback.print_exc()
```
